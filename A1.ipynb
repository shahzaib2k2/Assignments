```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
```

```python
path = 'C:\\Users\\engrs\\jsrt_metadata.csv'
df = pd.read_csv("jsrt_metadata.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['diagnosis'].replace(np.nan, df['diagnosis'].value_counts().idxmax(), inplace=True)
```

```python
df['size'].replace(np.nan, df['size'].astype(float).mean(), inplace=True)
```

```python
df['position'].replace(np.nan, df['position'].value_counts().idxmax(), inplace=True)
```

```python
df['y'].replace(np.nan, df['y'].astype(float).mean(), inplace=True)
```

```python
df['x'].replace(np.nan, df['x'].astype(float).mean(), inplace=True)
```

```python
df['subtlety'].replace(np.nan, df['subtlety'].astype(float).mean(), inplace=True)
```

```python
#Question:1
df['diagnosis'].value_counts() 
plt.pyplot.xlabel("Disease")
plt.pyplot.ylabel("Patients")
plt.pyplot.hist(df['diagnosis'], color='red',ec='blue')
```

```python
#Question:2
for i in list(df['diagnosis'].unique()):
    plt.pyplot.figure()
    plt.pyplot.pie(df[df['diagnosis'] == i]['state'].value_counts(),labels=list(df[df['diagnosis'] == i]['state'].value_counts().keys()),)
    df[df['diagnosis'] == i]['state'].value_counts()
```

```python
#Question:3
for i in list(df['diagnosis'].unique()):
    plt.pyplot.figure()
    plt.pyplot.pie(df[df['diagnosis'] == i]['gender'].value_counts(),labels=list(df[df['diagnosis'] == i]['gender'].value_counts().keys()),)
    df[df['diagnosis'] == i]['gender'].value_counts()
```
